### Using Conditionals to Filter Rows and Columns

For ex:- We want to look at our survey data and only look at people who know python - We can filter that out

```python
people = {
    "first": ["Jatin", 'Jane', 'John'], 
    "last": ["Dhiman", 'Doe', 'Doe'], 
    "email": ["JatinDhiman@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

Filter criteria or kind of mask - like person having last name Doe

```python
# Filter mask
df['last'] == 'Doe'

# Filter returns a series of true false values
```

To get the dataframe - Couple of ways

```python
# Need to assign the filter mask to a variable
filt = (df['last'] == 'Doe')
```

1st Method:

```python
df[filt]
#df[df['last'] == 'Doe']
```

2nd Method:

```python
# Use loc( by label) indexer (preferable)
df.loc[filt]
```

```python
df.loc[filt,'email']
```

Using AND and OR Operator to filter out data

```python
filt = ((df['last'] == 'Doe') & (df['first'] == 'John'))

df.loc[filt,'email']
```

```python
filt = ((df['last'] == 'Dhiman') | (df['first'] == 'John'))

df.loc[filt,'email']
```

Opposite of filter mask - Negate filter using ~

```python
filt = ((df['last'] == 'Dhiman') | (df['first'] == 'John'))

df.loc[~filt,'email']
```
